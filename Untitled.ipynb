```python
#parameters

core = ['Ting-Lu', "Iron Hands", "Amoonguss", "Flutter Mane", "Gyarados", "Chi-Yu"]
n = 10
bayesian = False
post_euic = False

```

```python
combo_list_dict = [{}, {}, {}, {}, {}, {}]

for index, row in mu_data.iterrows():
    tourney = row['Tourney']
    winner = row['Winner']
    loser = row['Loser']
    
    changeDict = {'Bloody X':'Giana Kaplan', 'Bambi M':'Benjamin Munger', 'Chuppa Cross IV':'Chuppa Cross', 
                  'TABLE 401 - Nils Schlatter':'Nils Schlatter', 'Tiago Calmon':'Tiago Salviano Calmon', 
                  'Blue Dee':'Giana Kaplan', 'May Gamelin':'Luke Gamelin'}
    if winner in changeDict:
        winner = changeDict[winner]
    if loser in changeDict:
        loser = changeDict[loser]
    
    try:
        winner_team = player_team_dict[tourney, winner]
        loser_team = player_team_dict[tourney, loser]
    except:
        print(winner, loser, tourney)
        continue
    if post_euic:
        if '''https://rk9.gg/pairings/euic-2022-vg''' == row['Url_Pairings']:
            break
    
    broken = False
    for mon in core:
        if mon not in winner_team:
            broken = True
            break
    if not broken:
        for i in range(6):
            #if 'shedinja' in loser_team:
            #    break
            temp = list(itertools.combinations(loser_team, i+1))
            for result in temp:
                #if result == ('exeggutor-alola',):
                #    print('loser', winner, winner_team, loser, loser_team, tourney)
                d = combo_list_dict[i]
                if result not in d:
                    if bayesian:
                        d[result] = [n/2, n/2]
                    else:
                        d[result] = [0,0]
                d[result][1] = d[result][1] + 1

                
                
    broken = False
    for mon in core:
        if mon not in loser_team:
            broken = True
            break
    if not broken:
        for i in range(6):
            #if 'shedinja' in winner_team:
            #    break
            temp = list(itertools.combinations(winner_team, i+1))
            for result in temp:
                #if result == ('exeggutor-alola',):
                #    print('winner', winner, winner_team, loser, loser_team, tourney)
                d = combo_list_dict[i]
                if result not in d:
                    if bayesian:
                        d[result] = [n/2, n/2]
                    else:
                        d[result] = [0,0]
                d[result][0] = d[result][0] + 1
```

```python
win_rate_list_of_lists = []
for d in combo_list_dict:
    pokemon_list = []
    win_rates = []
    for key in d:
        if bayesian == False:
            if d[key][0] + d[key][1] < n:
                continue
        pokemon_list.append(key)
        win_rates.append( d[key][0] / (d[key][0] + d[key][1])  )
    if len(win_rates) > 1:
        result1, result2 = zip(*sorted(zip(win_rates, pokemon_list), reverse = True, key=lambda x: x[0]))
    else:
        result1, result2 = win_rates, pokemon_list
    win_rate_list_of_lists.append([result2, result1])
```

```python
#create dictionaries
bay_game_constant = 3 #use 0 to remove
bay_game_constant2 = 0 #this one is for 2:1 data

one_to_zero_dict = {} #key is one pokemon, value is list of wins and losses
one_to_one_dict = {} #key is tuple of two pokemon, list is tuple of total wins and losses for first mon
two_to_one_dict = {} #key is tuple of three pokemon, list is tuple of total wins and losses for first two vs 3rd
win_diff_dict = {} #pokemon, +- wins
```

```python
#iterate through and fill dictionaries
for index, row in mu_data.iterrows():
    tourney = row['Tourney']
    winner = row['Winner']
    loser = row['Loser']
    
    changeDict = {'Bloody X':'Giana Kaplan', 'Bambi M':'Benjamin Munger', 'Chuppa Cross IV':'Chuppa Cross', 
                  'TABLE 401 - Nils Schlatter':'Nils Schlatter', 'Tiago Calmon':'Tiago Salviano Calmon', 
                  'Blue Dee':'Giana Kaplan', 'May Gamelin':'Luke Gamelin'}

    if winner in changeDict:
        winner = changeDict[winner]
    if loser in changeDict:
        loser = changeDict[loser]
    try:
        winner_team = player_team_dict[tourney, winner]
        loser_team = player_team_dict[tourney, loser]
    except:
        print(tourney, winner, "|", loser)
        continue
    
    list_of_teams_i_care_about = []
    #if 'amoonguss' in winner_team and 'arcanine' in winner_team:
    #    print(winner_team)
 
    
    for i in range(len(winner_team)):
        if 'Maushold' in winner_team[i]:
            winner_team[i] = 'Maushold'
        if 'Gastrodon' in winner_team[i]:
            winner_team[i] = 'Gastrodon'
        if 'Squawkabilly' in winner_team[i]:
            winner_team[i] = 'Squawkabilly'
        if 'Oricorio' in winner_team[i]:
            winner_team[i] = 'Oricorio'
        if 'Toxtricity' in winner_team[i]:
            winner_team[i] = 'Toxtricity'
        if 'Tatsugiri' in winner_team[i]:
            winner_team[i] = 'Tatsugiri'
        if winner_team[i] not in win_diff_dict:
            win_diff_dict[winner_team[i]] = 0
        win_diff_dict[winner_team[i]] += 1

    for i in range(len(loser_team)):
        if 'Maushold' in loser_team[i]:
            loser_team[i]= 'Maushold'
        if 'Gastrodon' in loser_team[i]:
            loser_team[i] = 'Gastrodon'
        if 'Squawkabilly' in loser_team[i]:
            loser_team[i]= 'Squawkabilly'
        if 'Oricorio' in loser_team[i]:
            loser_team[i] = 'Oricorio'
        if 'Toxtricity' in loser_team[i]:
            loser_team[i]= 'Toxtricity'
        if 'Tatsugiri' in loser_team[i]:
            loser_team[i] = 'Tatsugiri'
        if loser_team[i] not in win_diff_dict:
            win_diff_dict[loser_team[i]] = 0
        win_diff_dict[loser_team[i]] -= 1
    
    first = True
    for pokemonW in winner_team:
        if pokemonW not in one_to_zero_dict:
            one_to_zero_dict[pokemonW] = [0, 0]
        one_to_zero_dict[pokemonW][0] += 1
        
        for pokemonL in loser_team:
            
            if first:
                if pokemonL not in one_to_zero_dict:
                    one_to_zero_dict[pokemonL] = [0, 0]
                one_to_zero_dict[pokemonL][1] += 1
                
            
            if (pokemonW, pokemonL) not in one_to_one_dict:
                one_to_one_dict[(pokemonW, pokemonL)] = [bay_game_constant, bay_game_constant]
            one_to_one_dict[(pokemonW, pokemonL)][0] += 1
            
            if (pokemonL, pokemonW) not in one_to_one_dict:
                one_to_one_dict[(pokemonL, pokemonW)] = [bay_game_constant, bay_game_constant]
            one_to_one_dict[(pokemonL, pokemonW)][1] += 1
            
            for pokemonW2 in winner_team:
                if pokemonW == pokemonW2:
                    continue
                if (pokemonW, pokemonW2, pokemonL) not in two_to_one_dict:
                    two_to_one_dict[(pokemonW, pokemonW2, pokemonL)] = [bay_game_constant2, bay_game_constant2]
                two_to_one_dict[(pokemonW, pokemonW2, pokemonL)][0] += 1
            
            for pokemonL2 in loser_team:
                if pokemonL == pokemonL2:
                    continue
                if (pokemonL, pokemonL2, pokemonW) not in two_to_one_dict:
                    two_to_one_dict[(pokemonL, pokemonL2, pokemonW)] = [bay_game_constant2, bay_game_constant2]
                two_to_one_dict[(pokemonL, pokemonL2, pokemonW)][1] += 1
        first = False
    
```

```python
def winRate(l):
    return l[0]/(l[0] + l[1])
```
